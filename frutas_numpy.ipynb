```python
import numpy as np
```

Importando os dados e criando um array sem a coluna 0, visto que é uma string


```python
url = 'https://raw.githubusercontent.com/alura-cursos/numpy/dados/apples_ts.csv'
```

```python
dados = np.loadtxt(url, delimiter = ',', usecols = np.arange(1,88,1))
```

Diferenciando uma lista do Python para um array do NumPy

```python

# cria uma lista
lista = [1, 2, 3, 4, 5]

# transforma a lista em um array Numpy
array = np.array(lista)

print("Lista: ", lista)
print("Array: ", array)
```

Dimensões e legibilidade do array

```python
dados.ndim    # 2D, visto que são linhas e colunas
dados.size    # 522 registros
dados.shape   # 6 linhas e 87 colunas
```

```python
dados_transposto = dados.T        # realiza a transposição de uma matriz
                                  # cada coluna representa uma cidade
dados_transposto
```

Primeiros anos de Moscow

Comparando o preço das maçãs em dois anos diferentes:

```python
np.allclose(moscowAnoUm, moscowAnoDois, 10)     # diferença entre os dois anos não é maior do que 10 em relação a tolerancia absoluta
```
