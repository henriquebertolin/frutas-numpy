Lidando com NaNs

```python
plt.plot(datas, kaliningrad)
kaliningrad
# é possível observar um espaço em branco logo no início do gráfico, que corresponde ao nan do array
```

```python
sum(np.isnan(kaliningrad))
```

```python
inds = np.where(np.isnan(kaliningrad))
inds
```

```python
for i in inds:
  if i > 0 and i < len(kaliningrad)-1:
    kaliningrad[i] = (kaliningrad[i-1] + kaliningrad[i+1]) / 2
  elif i == 0:
    kaliningrad[i] = kaliningrad[i+1]
  else:
    kaliningrad[i] = kaliningrad[i-1]
```

Equação para se obter o coeficiente angular:

image.png

a = coeficiente angular <br>
n = numero de elementos<br>
y = moscow<br>
x = datas

```python
y= moscow
x = datas
n = np.size(y)
a = (n * np.sum(x*y) - np.sum(x) * np.sum(y)) / (n*np.sum(x**2) - np.sum(x)**2)
a
```

Equação para encontrar o coeficiente linear:

image.png

```python
b = np.mean(y) - a * np.mean(x)
b
```

```python
y = a * x + b
np.linalg.norm(moscow - y)
```

```python
plt.plot(datas, moscow)
plt.plot(x,y)
```

Para que serve a regressão linear e como utiliza-lá?

```python
# Calcular valores intermediários dentro e fora da reta:
plt.plot(datas, moscow)
plt.plot(x,y)
plt.plot(41.5, 41.5*a+b, '*r')
```
